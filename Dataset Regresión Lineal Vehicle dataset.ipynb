# Preprocesamiento de datos

```python
# Eliminar la columna Car_Name
car_data.drop(labels='Car_Name',axis= 1, inplace = True)
```

```python
# Comprobar
car_data.head()
```

```python
# Convertir las variables categoricas a numericas
car_data = pd.get_dummies(car_data)

# Selecciona solo las columnas booleanas (dummies) para convertirlas a enteros
bool_columns = car_data.select_dtypes(include='bool').columns

# Convierte las columnas booleanas a enteros (0 y 1)
car_data[bool_columns] = car_data[bool_columns].astype(int)

# Verifica el resultado
car_data.head()
```

# Selección de características y  División del dataset en Train y Test

```python
# crear X (columnas independientes) e y (columna dependiente o objetivo)
y = car_data['Selling_Price']
X = car_data.drop('Selling_Price',axis=1)
```

```python
from sklearn.model_selection import train_test_split

# Dividir el dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

# Entrenar el modelo configurando los diferentes hiperparámetros

```python
# Implementación de LinearRegressionModel
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# # Realizar predicciones en el conjunto de prueba usando el modelo entrenado
prediction = lr.predict(X_test)
# Puntuación de entrenamiento
print("Training score: ", lr.score(X_train, y_train))
# Puntación de la prueba
print("test score: ", lr.score(X_test,y_test)) 
```

# Evaluar el desempeño del modelo

```python
from sklearn import metrics

print('MAE : ', metrics.mean_absolute_error(y_test, prediction))
print('MSE : ', metrics.mean_squared_error(y_test, prediction))
print('RMSE : ', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
print('R squared : ', metrics.r2_score(y_test, prediction))
```

#  Realizar las diferentes gráficas que permitan visualizar los resultados del modelo

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, prediction, alpha=0.7)
plt.xlabel('Valores Reales')
plt.ylabel('Valores Predichos')
plt.title('Valores Reales vs. Predichos')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.show()
```

# Interpretar, analizar y documentar los resultados obtenidos

Del desempeño del modelo se puede interpretrar:

- Coeficiente de determinación (R^2):  El R^2 mide la proporción de la variabilidad en la variable objetivo (Selling_Price) que es explicada por el modelo. Un valor de 0.8770 significa que el 87.7% de la variabilidad en los precios de venta de los coches se explica por las características o variables independientes (año, precio actual, tipo de combustible, tipo de vendedor, kms recorrido, entre otros.) incluidas en el modelo.
 - 
Error Absoluto Medio (MAE): Mide el error promedio entre los valores predichos y los reales, en este caso, en promedio, el modelo tiene un error de   aproximadamente 1.23 lo que significa que en promedio las predicciones del modelo están a 1.23 unidades del valor real.  
- 
Error Cuadrático Medio (MSE): El MSE mide el error promedio al cuadrado entre las predicciones y los valores reales. Penaliza más los errores grandes que el MAE, ya que los errores se elevan al cuadrado. En este caso, el MSE es 3.38, lo que sugiere que hay algunos errores más grandes en la       s prediccionesdes.


Del grafico de dispersión se puede interpretar:

- El gráfico indica que el modelo está realizando un trabajo aceptable en la predicción de los valores, ya que muchos puntos están cerca de la línea de referencia. Los puntos alejados que se observan en el grafico pueden afectar las métricas como el RMSE o el MSE ya que estas métricas son sensibles a los datos atípicos. 

- En conclusión, el modelo tiene un buen rendimiento en general, dado que el r² es alto (0.8770), lo que significa que las predicciones se ajustan bien a los datos reales. el mae y el rmse están en el mismo rango (1.2 - 1.8), lo que sugiere que el error en las predicciones es relativamente constante y no hay muchos errores extremadamente grandes.
