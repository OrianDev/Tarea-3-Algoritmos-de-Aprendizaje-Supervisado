# Preprocesamiento de datos y Selección de características

-Convertir las variables categoricas a binarias con la tecnica one-hot 

-Estandarizar las variables numericas no binarias

```python
def onehot_encode(df, column_dict):
    df = df.copy()
    for column, prefix in column_dict.items():
        dummies = pd.get_dummies(df[column], prefix=prefix) # Realiza la codificación one-hot
        df = pd.concat([df, dummies], axis=1) # Agrega las columnas codificadas al dataframe
        df = df.drop(column, axis=1) # Elimina la columna original
    return df 
```

```python
def preprocess_inputs(df, scaler):
    df = df.copy()
    
    # Definir las variables categoricas que se deben codificar con one-hot
    nominal_features = ['cp', 'slope', 'thal']
    df = onehot_encode(df, dict(zip(nominal_features, ['CP', 'SL', 'TH'])))
    
     # Separar el dataframe en la variable objetivo 'y' y las variables independientes 'X'
    y = df['condition'].copy()
    X = df.drop('condition', axis=1).copy()
    
    # Estandarizar las variables numericas independientes de 'X' utilizando el escalador proporcionado
    # Esto convierte 'X' a una distribución con media 0 y desviación estándar 1
    X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
    
    return X, y
```

# División del dataset en Train y Test

```python
#La función train_test_split crea una división de un conjunto de datos en dos bloques uno de entrenamiento y otro de prueba (train and test)
# Porcentaje de datos para entrenamiento (train) = 80% , y para prueba (test) = 20%
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=0)
```

# Entrenar el modelo configurando los diferentes hiperparámetros

```python
# Construir el modelo de regresión logística 
modelo_regresion = LogisticRegression()
modelo_regresion.fit(X_train, y_train)

# Crea el modelo de prediccion
Prediccion = modelo_regresion.predict(X_test)
```

# Evaluar el desempeño del modelo en el conjunto de Test con métricas como precisión, recall, F1-score, etc
